```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from scipy.stats import norm
# Generate 1000 normal random integers with specified mean and standard deviation.
draw = norm.ppf(np.random.random(100000),loc=1000,scale=200).astype(int) # mean=1000, std=200
sns.distplot(draw)
```

```python
l =[]
for i in draw:
    l.append((i-1000)/(200))
```

```python
a = np.array(l)
```

```python
print(a.mean())
```

```python
print(np.round(a.mean()))
```

```python
print(a.std())
print(np.round(a.std()))
```

```python
sns.distplot(a)
```
